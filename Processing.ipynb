```python
import urllib.request, xml.etree.ElementTree as ET

response = urllib.request.urlopen("https://www.ecb.europa.eu/stats/policy_and_exchange_rates/euro_reference_exchange_rates/html/usd.xml")
data = response.read()
root = ET.fromstring(data)
```

```python
eur_usd = []
for child in root:
    if child.tag == '{http://www.ecb.europa.eu/vocabulary/stats/exr/1}DataSet':
        for i in child:
            for j in i:
                eur_usd.append([j.attrib.get('TIME_PERIOD'), j.attrib.get('OBS_VALUE')])
```

```python
eur_usd = pd.DataFrame(eur_usd, columns = ['Date', 'USD_EUR'])
eur_usd['Date'] = pd.to_datetime(eur_usd['Date'])
```

```python
def merging(json_passes: list, csv_passes: list, output_pass: str):
    import json
    import pandas as pd
    import datetime
    people = []
    assets = []
    for num, i in enumerate(json_passes):
        with open(i) as f:
            person=json.load(f)
        person_lst = []
        for j in person:
            person_lst.append([person[j]['created_at'], person[j]['user']['name'], person[j]['text']])
        person = pd.DataFrame(person_lst, columns = ['Date', 'Name', 'Tweets'])
        person['Date'] = pd.to_datetime(person['Date'])
        person['Date'] = person['Date'].dt.strftime('%Y-%m-%d')
        person['Date'] = pd.to_datetime(person['Date'])
        people.append(person)
        
    for l in csv_passes:
        asset = pd.read_csv(l) 
        asset['Date'] = pd.to_datetime(asset['Date'])
        assets.append(asset)

    a = people[0]
    b = assets[0]
    for i in range(1,len(people)):
        a = a.append(people[i])
    for i in range(1,len(assets)):
        b = b.merge(assets[i], how='outer')
        
    start_date = max(min(a['Date']), min(b['Date']))
    tweet_dates = a[a['Date'] >= start_date]['Date']
    asset_dates = b[(b['Date'] >= min(tweet_dates)) & (b['Date'] <= max(tweet_dates))]['Date']
    missing_dates = pd.concat([asset_dates,tweet_dates]).drop_duplicates(keep=False)
    missing_assets = b[b['Date'].isin(missing_dates)]
    for clm in a.columns[1:]:
        missing_assets[clm] = None  
        
    total = a.merge(b)
    if max(a['Date']) != max(total['Date']):
        add_clm = []
        for _ in a.iloc[0, :]:
            add_clm.append(_)
        for _ in range(len(total.columns) - len(a.columns)):
            add_clm.append(None)
        add_clm = pd.DataFrame([add_clm])
        add_clm.columns = total.columns
        total = total.append(add_clm)

    total = total.append(missing_assets)
    total.sort_values('Date', ascending=False, inplace=True)
    total.to_csv(output_pass, index = False)
    return total.reset_index(drop = True) 
```

```python
def merging_custom(json_passes: list, csv_passes: list, \
        output_pass: str, additional_keys = [], columns_to_include = []):
    import json
    import pandas as pd
    import datetime
    people = []
    assets = []
    for i in json_passes:
        with open(i) as f:
            person=json.load(f)
        person_lst = []
        for j in person:
            person_lst.append([person[j]['created_at'], person[j]['user']['name'], person[j]['text']] +\
                               [person[j].get(key, None) for key in additional_keys if additional_keys != []])
        person = pd.DataFrame(person_lst, columns = ['Date', 'Name', 'Tweets']+\
                                                     [key for key in additional_keys if additional_keys != []])
        person['Date'] = pd.to_datetime(person['Date'])
        person['Date'] = person['Date'].dt.strftime('%Y-%m-%d')
        person['Date'] = pd.to_datetime(person['Date'])
        people.append(person)
        
    for k in csv_passes:
        asset = pd.read_csv(k) 
        if columns_to_include != []: 
            if 'Date' not in columns_to_include:
                columns_to_include.append('Date')
            asset = asset[[_ for _ in columns_to_include if _ in asset.columns]]
        asset['Date'] = pd.to_datetime(asset['Date'])
        assets.append(asset)
    
    a = people[0]
    b = assets[0]
    for i in range(1,len(people)):
        a = a.append(people[i])
    for i in range(1,len(assets)):
        b = b.merge(assets[i], how='outer')
        
    start_date = max(min(a['Date']), min(b['Date']))
    tweet_dates = a[a['Date'] >= start_date]['Date']
    asset_dates = b[(b['Date'] >= min(tweet_dates)) & (b['Date'] <= max(tweet_dates))]['Date']
    missing_dates = pd.concat([asset_dates,tweet_dates]).drop_duplicates(keep=False)
    missing_assets = b[b['Date'].isin(missing_dates)]
    for clm in a.columns[1:]:
        missing_assets[clm] = None

    total = a.merge(b)
    
    if max(a['Date']) != max(total['Date']):
        add_clm = []
        for _ in a.iloc[0, :]:
            add_clm.append(_)
        for _ in range(len(total.columns) - len(a.columns)):
            add_clm.append(None)
        add_clm = pd.DataFrame([add_clm])
        add_clm.columns = total.columns
        total = total.append(add_clm)
        
    total = total.append(missing_assets)
    total.sort_values('Date', ascending=False, inplace=True)    
    total.to_csv(output_pass, index = False)
    return total 
```
